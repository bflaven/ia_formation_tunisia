```python
# @title Chargement du Modèle NLLB et du Tokenizer

def charger_modele_et_tokenizer():
    """
    # Charge le modèle NLLB et le tokenizer
    # Nous utilisons la version NLLB-200-distilled-600M pour un équilibre entre performance et rapidité
    """
    print("Chargement du modèle NLLB et du tokenizer en cours...")

    # Point de contrôle du modèle - version distillée pour plus de rapidité
    model_name = "facebook/nllb-200-distilled-600M"

    # Chargement du tokenizer
    tokenizer = AutoTokenizer.from_pretrained(model_name)

    # Chargement du modèle
    model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

    # Si un GPU est disponible, on utilise CUDA pour accélérer les calculs
    device = "cuda" if torch.cuda.is_available() else "cpu"
    model = model.to(device)

    print(f"Modèle chargé avec succès! Utilisation de: {device}")

    return model, tokenizer, device
```

```python
# @title Modèle de Traduction Simple

def traduire_texte_simple(source_text, langue_source, langue_cible):
    """
    # Version simplifiée de la traduction pour un usage direct
    """
    print("Chargement du modèle et du tokenizer...")

    # Chargement du modèle et du tokenizer
    model_name = "facebook/nllb-200-distilled-600M"
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

    device = "cuda" if torch.cuda.is_available() else "cpu"
    model.to(device)

    print(f"Modèle chargé sur {device}")

    # NLLB utilise des tokens spéciaux pour les langues
    # On prépare le texte d'entrée
    print("Préparation de la traduction...")
    inputs = tokenizer(source_text, return_tensors="pt").to(device)

    # La traduction avec NLLB se fait avec un token spécial pour la langue cible
    # On va utiliser directement une approche qui fonctionne de manière fiable

    # La méthode simplifiée pour NLLB est d'utiliser le préfixe de la langue cible
    target_token = f"___{langue_cible}___"
    forced_bos_token_id = tokenizer.convert_tokens_to_ids(target_token)

    # Si le token n'est pas trouvé, on essaie une autre approche
    if forced_bos_token_id == tokenizer.unk_token_id:
        print("Méthode alternative pour la langue cible...")
        try:
            # Pour les versions plus récentes de transformers
            if hasattr(tokenizer, 'lang_code_to_id'):
                forced_bos_token_id = tokenizer.lang_code_to_id.get(langue_cible)
            else:
                # Méthode différente sur certaines versions
                token_lang = tokenizer.convert_tokens_to_ids(langue_cible)
                forced_bos_token_id = token_lang if token_lang != tokenizer.unk_token_id else None
        except:
            print("Avertissement: Impossible de définir le token de langue cible de manière standard.")
            forced_bos_token_id = None

    print("Traduction en cours...")

    # Paramètres de génération
    generate_kwargs = {
        "max_length": 400,
        "num_beams": 5,
        "early_stopping": True
    }

    # Si nous avons un token de langue cible valide, on l'utilise
    if forced_bos_token_id is not None:
        generate_kwargs["forced_bos_token_id"] = forced_bos_token_id

    # Génération de la traduction
    translated_tokens = model.generate(**inputs, **generate_kwargs)

    # Décodage du résultat
    translation = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]

    print("Traduction terminée!")

    return translation
```

```python
# @title Interface de Traduction

# Exemple de texte source en français
default_text = """
Thirty Tunisian athletes will be participating in the Paris Paralympics – 13 women and 17 men across four disciplines.
"""
# content from article https://www.france24.com/en/video/20240823-tunisian-athletes-gear-up-for-paris-paralympics

# Zone de texte pour saisir le texte source
from ipywidgets import widgets, Layout

# Zone de texte pour saisir le texte source
source_text_widget = widgets.Textarea(
    value=default_text,
    placeholder='Entrez le texte à traduire ici',
    description='Texte source:',
    layout=Layout(width='100%', height='150px')
)

# Obtention de la liste des langues
print("Chargement de la liste des langues disponibles...")
langues_disponibles = afficher_langues_disponibles()

# Création des listes pour les menus déroulants
options_langues = sorted([(v, k) for k, v in langues_disponibles.items()])

# Menu déroulant pour choisir la langue source
langue_source_widget = widgets.Dropdown(
    options=options_langues,
    # value='fra_Latn',  # Français par défaut
    value='eng_Latn',  # Anglais par défaut

    description='Langue source:',
)

# Menu déroulant pour choisir la langue cible
langue_cible_widget = widgets.Dropdown(
    options=options_langues,
    # value='eng_Latn',  # Anglais par défaut
    value='fra_Latn',  # Français par défaut

    description='Langue cible:',
)

# Bouton pour lancer la traduction
bouton_traduire = widgets.Button(
    description='Traduire',
    button_style='primary',
    tooltip='Cliquez pour traduire'
)

# Zone d'affichage du résultat
resultat_widget = widgets.Textarea(
    placeholder='Le résultat de la traduction apparaîtra ici',
    description='Résultat:',
    layout=Layout(width='100%', height='150px'),
    disabled=True
)

# Fonction exécutée lorsqu'on clique sur le bouton
def on_bouton_traduire_click(b):
    # Récupération des valeurs
    texte = source_text_widget.value
    langue_source = langue_source_widget.value
    langue_cible = langue_cible_widget.value

    # Vérification que le texte n'est pas vide
    if not texte.strip():
        resultat_widget.value = "Erreur: Le texte source est vide!"
        return

    # Affichage d'un message d'attente
    resultat_widget.value = "Traduction en cours... Veuillez patienter."

    try:
        # Utilisation de la méthode simplifiée pour éviter les erreurs
        texte_traduit = traduire_texte_simple(texte, langue_source, langue_cible)

        # Affichage du résultat
        resultat_widget.value = texte_traduit
    except Exception as e:
        # En cas d'erreur, on l'affiche
        resultat_widget.value = f"Erreur lors de la traduction: {str(e)}"

# Association du bouton à la fonction
bouton_traduire.on_click(on_bouton_traduire_click)

# Affichage des widgets
display(widgets.HTML("<h3>Traduction avec NLLB (No Language Left Behind)</h3>"))
display(widgets.VBox([
    widgets.HTML("<h4>1. Saisissez le texte à traduire:</h4>"),
    source_text_widget,
    widgets.HTML("<h4>2. Sélectionnez les langues:</h4>"),
    widgets.HBox([langue_source_widget, langue_cible_widget]),
    widgets.HTML("<h4>3. Lancez la traduction:</h4>"),
    bouton_traduire,
    widgets.HTML("<h4>4. Résultat de la traduction:</h4>"),
    resultat_widget
]))
```
